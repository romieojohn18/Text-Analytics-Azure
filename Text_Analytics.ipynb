```python
import torch
from transformers import AutoModelForTokenClassification, AutoTokenizer

# 1. Load the pre-trained GatorTron model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract-fulltext")
model = AutoModelForTokenClassification.from_pretrained("microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract-fulltext")

# 2. Define text to process
text = "The patient was prescribed 500mg of Metformin twice daily for their diabetes."

# 3. Tokenize the text
inputs = tokenizer(text, return_tensors="pt")

# 4. Predict entity labels for each token
outputs = model(**inputs)
predictions = torch.argmax(outputs.logits, dim=2)

# Print predictions and labels
print("Predictions:", predictions)
print("Labels:", [model.config.id2label[label.item()] for label in predictions[0]])

# Print the mapping between label indices and label names
print("Label mapping:", model.config.id2label)

# Extracted medicine names based on predicted entity labels
medicine_names = []
current_entity = []

for token, prediction in zip(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0]), predictions[0]):
    label = tokenizer.convert_ids_to_tokens(prediction.item())

    if label == 'B-CHEMICAL':
        current_entity.append(token.replace("##", ""))  # Remove '##' in case of subword tokens
    elif label == 'I-CHEMICAL' and current_entity:
        current_entity.append(token.replace("##", ""))
    elif current_entity:
        medicine_names.append(" ".join(current_entity))
        current_entity = []

# Add the last entity if any
if current_entity:
    medicine_names.append(" ".join(current_entity))

# Print the extracted medicine names
print("Extracted medicine names:", medicine_names)
```

```python
def gather_information():
    # Patient Information
    first_name = input("Enter patient's first name: ")
    last_name = input("Enter patient's last name: ")
    dob = input("Enter patient's date of birth (YYYY-MM-DD): ")
    examination_date = input("Enter examination date (YYYY-MM-DD): ")

    # Medical History
    medical_history = input("Enter patient's medical history: ")

    # Radiology Section
    radiology = input("Enter radiology results: ")

    # Testing Section
    testing = input("Enter testing results: ")

    return {
        "first_name": first_name,
        "last_name": last_name,
        "dob": dob,
        "examination_date": examination_date,
        "medical_history": medical_history,
        "radiology": radiology,
        "testing": testing,
    }

def generate_report(patient_info):
    report = f"""
    Medical Examination Report
    ---------------------------
    Patient Information:
    First Name: {patient_info['first_name']}
    Last Name: {patient_info['last_name']}
    Date of Birth: {patient_info['dob']}
    Examination Date: {patient_info['examination_date']}

    Medical History:
    {patient_info['medical_history']}

    Radiology Results:
    {patient_info['radiology']}

    Testing Results:
    {patient_info['testing']}
    """
    return report

def save_report(report, filename="medical_examination_report.txt"):
    with open(filename, "w") as file:
        file.write(report)
    print(f"Report saved as {filename}")

def main():
    patient_info = gather_information()
    report = generate_report(patient_info)
    print("\nGenerated Report:\n", report)
    save_report(report)

if __name__ == "__main__":
    main()
```

```python
# Function to print a simple text-based table
def print_table(headers, data):
    # Determine column widths by finding the maximum length of data in each column
    col_widths = [max(len(str(value)) for value in col) for col in zip(*data, headers)]
    # Create a format string using the column widths
    row_format = "|".join([" {:<" + str(width) + "} " for width in col_widths])
    # Print the header
    print(row_format.format(*headers))
    print("-" * sum(col_widths) + "-" * (3 * len(headers) - 1))  # Print a divider
    # Print each row of data
    for row in data:
        print(row_format.format(*row))

# Repeated Medicines on the Same Date
headers_repeated_dates = ['Date', 'Medicine', 'Times Given']
data_repeated_dates = [
    ['09/11/2017', 'Concor Cor 25 Mg Tab', 'Twice'],
    ['09/11/2017', 'Pantocid 40 Mg Inj', 'Twice'],
    ['11/11/2017', 'Syringe 1 MI Bd 40 Units', 'Twice'],
    ['12/11/2017', 'Eptus 25 Mg Tab', 'Twice']
]
print("Repeated Medicines on the Same Date:")
print_table(headers_repeated_dates, data_repeated_dates)
print("\n")  # Adding a newline for better readability

# List of Repeated Medicines in the File
headers_repeated_medicines = ['No.', 'Medicine']
data_repeated_medicines = [
    [1, 'Concor Cor 25 Mg Tab'],
    [2, 'Ecosprin 75 Mg Tab'],
    [3, 'Eptus 25 Mg Tab'],
    [4, 'Pantocid 40 Mg Inj'],
    [5, 'Atorva 40 Mg Tab'],
    [6, 'Syringe 1 MI Bd 40 Units']
]
print("List of Repeated Medicines in the File:")
print_table(headers_repeated_medicines, data_repeated_medicines)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download required NLTK data (if not already done)
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    # Tokenize and convert to lowercase
    tokens = word_tokenize(text.lower())

    # Get English stopwords
    stop_words = set(stopwords.words('english'))

    # Filter out tokens that are not alphanumeric and are in the list of stopwords
    cleaned_tokens = [token for token in tokens if token.isalnum() and token not in stop_words]

    # Join the cleaned tokens into a string
    cleaned_text = ' '.join(cleaned_tokens)

    return cleaned_text

pdf_path ="sample.pdf"
extracted_text = extract_text_from_pdf(pdf_path)
preprocessed_text = preprocess_text(extracted_text)
print(preprocessed_text)
```
